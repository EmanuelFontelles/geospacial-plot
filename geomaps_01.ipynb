```python
import plotly as py
import plotly.graph_objs as go
import pandas as pd
```

```python
df = pd.read_csv("2017.csv")
```

```python
df.head(10)
```

```python
data = dict (
    type = 'choropleth',
    locations = df['Country'],
    locationmode='country names',
    colorscale = 'Blackbody',
    z=df['Happiness.Score'])
```

```python
map1 = go.Figure(data=[data])
# py.offline.plot(map1)
py.plotly.iplot(map1)
```

```python
data = dict (
    type = 'choropleth',
    locations = ['AZ','CA','VT'],
    locationmode='USA-states',
    colorscale = 'Viridis',
    z=[10,20,30])

lyt = dict(geo=dict(scope='usa'))
```

```python
map2 = go.Figure(data=[data], layout = lyt)
py.plotly.iplot(map2)
```
